```python
cols = ['BLOWING SAND/SNOW', 'CLEAR', 'CLOUDY', 'FOG','OTHER_WEATHER',
        'RAIN', 'SEVERE CROSSWINDS', 'SLEET/HAIL',
       'SNOW_WEATHER', 'UNKNOWN_WEATHER', 'DARK-LIGHTED', 'DARK-NOT LIGHTED',
       'DARK-UNKNOWN LIGHTING', 'DAWN', 'DAYLIGHT', 'DUSK',
       'UNKNOWN_LIGHT', 'INATTENTION', 'SPEEDING', 'CITY STREET',
       'COUNTY ROAD', 'FARM TO MARKET', 'INTERSTATE', 'OTHER ROADS',
       'TOLL BRIDGES', 'TOLLWAY', 'US & STATE HIGHWAY',
        'ALCOHOL', 'ALCOHOL, DRUG', 'DRUG', 'MARIJUANA',
       'MEDICINAL', 'NO INTOXICATION','POPULATION','OTHER_LIGHT', 
       'STATE_Colorado', 'STATE_Oklahoma','YEAR_2019', 'YEAR_2020']
```

```python
X = data[cols]
```

```python
X.head()
```

```python
X.isnull().sum()
```

```python
y = data['LOG_INJURY_RATE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=.20)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
rf.fit(X_train_scaled,y_train)
```

```python
y_pred = rf.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
r2score = r2_score(y_test,y_pred)
```

```python
r2score
```

```python
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.model_selection import cross_val_score
```

```python
pipeline = make_pipeline(scaler,rf)
```

```python
cv_scores = cross_val_score(pipeline,X_train_scaled,y_train,cv=5,scoring='r2')
```

```python
cv_avg = np.average(cv_scores)
cv_avg
```
