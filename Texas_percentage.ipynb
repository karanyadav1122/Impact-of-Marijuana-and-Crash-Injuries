```python
total_counts_by_county = counts_2021_df.groupby('COUNTY')['COUNT'].sum()
total_counts_by_county
```

```python
SPEEDING = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '1'].groupby('COUNTY')['COUNT'].sum()
INATTENTION = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '2'].groupby('COUNTY')['COUNT'].sum()
LANE_CHNG = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '3'].groupby('COUNTY')['COUNT'].sum()
TAILGATING = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '4'].groupby('COUNTY')['COUNT'].sum()
VIOLATION = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '5'].groupby('COUNTY')['COUNT'].sum()
UNSAFE_BACKING = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '6'].groupby('COUNTY')['COUNT'].sum()
IMPAIRMENT = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '7'].groupby('COUNTY')['COUNT'].sum()
TURNING = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '8'].groupby('COUNTY')['COUNT'].sum()
TECH_PARK = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '9'].groupby('COUNTY')['COUNT'].sum()
NO_FAULT = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '10'].groupby('COUNTY')['COUNT'].sum()
KNW_HMC_FACTOR = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '11'].groupby('COUNTY')['COUNT'].sum()
UNW_HMC_FACTOR = counts_2021_df[counts_2021_df['HUMAN_FACTOR'] == '12'].groupby('COUNTY')['COUNT'].sum()
```

```python
# Calculate percentages for each human factor category
speeding_percentages = (SPEEDING / total_counts_by_county) * 100
inattention_percentages = (INATTENTION / total_counts_by_county) * 100
lane_change_percentages = (LANE_CHNG / total_counts_by_county) * 100
tailgating_percentages = (TAILGATING / total_counts_by_county) * 100
violation_percentages = (VIOLATION / total_counts_by_county) * 100
unsafe_backing_percentages = (UNSAFE_BACKING / total_counts_by_county) * 100
impairment_percentages = (IMPAIRMENT / total_counts_by_county) * 100
turning_percentages = (TURNING / total_counts_by_county) * 100
tech_park_percentages = (TECH_PARK / total_counts_by_county) * 100
no_fault_percentages = (NO_FAULT / total_counts_by_county) * 100
known_human_factor_percentages = (KNW_HMC_FACTOR / total_counts_by_county) * 100
unknown_human_factor_percentages = (UNW_HMC_FACTOR / total_counts_by_county) * 100

```

```python
# Concatenate the human factor percentage Series into a DataFrame
human_factor_percentages_df = pd.concat([
    speeding_percentages, 
    inattention_percentages, 
    lane_change_percentages, 
    tailgating_percentages, 
    violation_percentages, 
    unsafe_backing_percentages, 
    impairment_percentages, 
    turning_percentages, 
    tech_park_percentages, 
    no_fault_percentages, 
    known_human_factor_percentages, 
    unknown_human_factor_percentages
], axis=1)

# Assign new column names corresponding to each human factor
human_factor_percentages_df.columns = [
    'SPEEDING_PERCENTAGE', 
    'INATTENTION_PERCENTAGE', 
    'LANE_CHANGE_PERCENTAGE', 
    'TAILGATING_PERCENTAGE', 
    'VIOLATION_PERCENTAGE', 
    'UNSAFE_BACKING_PERCENTAGE', 
    'IMPAIRMENT_PERCENTAGE', 
    'TURNING_PERCENTAGE', 
    'TECH_PARK_PERCENTAGE', 
    'NO_FAULT_PERCENTAGE', 
    'KNOWN_HUMAN_FACTOR_PERCENTAGE', 
    'UNKNOWN_HUMAN_FACTOR_PERCENTAGE'
]

# Reset the index to turn the county index into a column
human_factor_percentages_df = human_factor_percentages_df.reset_index().rename(columns={'index': 'COUNTY'})

# Display the DataFrame to verify everything looks correct
human_factor_percentages_df
```

```python
def process_lighting_data(df, mapping):
    # Map the lighting codes to descriptive names
    df['LIGHTING_DESC'] = df['LIGHTING'].map(mapping).astype(str)
    
    # Group by COUNTY and LIGHTING_DESC, then count occurrences
    grouped_lighting = df.groupby(['COUNTY', 'LIGHTING_DESC']).size().unstack(fill_value=0)
    
    # Calculate total counts by county to normalize the data
    total_counts_by_county_lighting = grouped_lighting.sum(axis=1)
    
    # Calculate percentages
    lighting_percentages_df = (grouped_lighting.div(total_counts_by_county_lighting, axis=0) * 100).fillna(0)
    lighting_percentages_df = lighting_percentages_df.reset_index()
    
    return lighting_percentages_df

# Apply the function to each yearly DataFrame
lighting_percentages_df_2019 = process_lighting_data(df_2019, lighting_conditions)
lighting_percentages_df_2020 = process_lighting_data(df_2020, lighting_conditions)
lighting_percentages_df_2021 = process_lighting_data(df_2021, lighting_conditions)
```
