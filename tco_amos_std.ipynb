```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/karan/tco_amos_std.csv')
data
```

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
y = data['INR']
y
```

```python
y = y.array.reshape(-1,1)
y
```

```python
y1 = scaler.fit_transform(y)
```

```python
y1
```

```python
y2 = pd.DataFrame(data= y1, columns=['INR_S'])
```

```python
y2
```

```python
data1 = pd.concat([data,y2],axis= 1)
```

```python
data1
```

```python
data1
```

```python
data1.columns
```

```python
data1.columns = data1.columns.str.replace(r"\(\'|\'\)", "")
```

```python
data1.columns = data1.columns.str.replace(r"\(\'|\'\)", "").str.replace(r",\s*", "")
```

```python
data1.columns = data1.columns.str.replace(r"\)\s*", "").str.replace(r"\(\'|\'\)", "")
```

```python
data1.columns = data1.columns.str.replace(r"\'", "")
```

```python
data1.to_csv('tco_amos_std.csv',index= False)
```
